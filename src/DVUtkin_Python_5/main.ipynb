## Задание 1
Создайте класс <b>Student</b>, который должен содержать:

+ Конструктор, принимающий имя и возраст студента;
+ Атрибуты <b>name</b> (публичный) и <b>age</b> (защищенный);
+ Метод <b>get_info()</b>, который возвращает строку: <b>"Студент {name}, возраст {age}"</b>.

```python
class Student:
    """Класс для обычного студента.
    
    Атрибуты:
        name (str): Имя студента.
        _age (int): Возраст студента.        
    """

    def __init__(self, name:str, age:int):
        """Конструктор класса Student.

        Args:
            name (str): Имя студента.
            age (int): Возраст студента.
        """
        self.name = name
        self._age = age

    def get_info(self):
        """Возвращает информацию о студенте.

        Returns:
            str: Строка формата "Студент {name}, возраст {age}"
        """
        return f'Студент {self.name}, возраст {self._age}'
```

```python
# Образец использования:
student1 = Student('Анна', 20)
print(student1.get_info())  # Должно вывести: "Студент Анна, возраст 20"
```

## Задание 2
Расширьте класс <b>Student</b> из первой задачи, <b>создав</b> класс <b>GraduateStudent</b> (аспирант), который:

+ Наследуется от <b>Student</b>;
+ Добавляет защищенный атрибут <b>_research_topic</b> (тема исследования);
+ Добавляет приватный атрибут <b>__publications</b> (количество публикаций);
+ Добавляет метод <b>add_publication()</b>, который увеличивает количество публикаций на 1;
+ Переопределяет метод <b>get_info()</b>, чтобы он включал тему исследования.


```python
class GraduateStudent(Student):
    """Класс для аспиранта, наследуюется от Student.
    
    Атрибуты:
        name (str): Имя студента.
        _age (int): Возраст студента.  
        _research_topic (str): Тема исследования.
        __publications (int): Количество публикаций.    
    """

    def __init__(self, name:str, age:int, research_topic:str):
        """Конструктор класса GraduateStudent.

        Args:
            name (str): Имя студента.
            age (int): Возраст студента.
            research_topic (str): Тема исследования.
        """
        super().__init__(name, age)
        self._research_topic = research_topic
        self.__publications = 0
    
    @property
    def publications(self):
        """Возвращает количество публикаций.

        Returns:
            int: Количество публикаций.
        """
        return self.__publications
    
    def add_publication(self):
        """Увеличивает количество публикаций на 1."""
        self.__publications += 1
    
    def get_info(self):
        """Возвращает информацию о студенте.

        Returns:
            str: Строка формата "Студент {name}, возраст {age}, тема: {research_topic}"
        """        
        return f'Студент {self.name}, возраст {self._age}, тема: {self._research_topic}'
```

```python
# Образец использования:
grad_student = GraduateStudent('Петр', 25, 'Искусственный интеллект')
print(grad_student.get_info())  # "Студент Петр, возраст 25, тема: Искусственный интеллект"

grad_student.add_publication()
grad_student.add_publication()
print(f'Студент {grad_student.name}, {grad_student.publications} публикаций')
```

## Задание 3
Создайте систему управления студентами:

+ Реализуйте класс <b>StudentManager</b> для работы с коллекцией студентов;
+ Добавьте магические методы в классы:

    + <b>\_\_str\_\_</b> и <b>\_\_repr\_\_</b> для красивого вывода;
    + <b>\_\_len\_\_</b> для StudentManager (возвращает количество студентов);
    + <b>\_\_getitem\_\_</b> для доступа к студентам по индексу;

+ Реализуйте полиморфизм - метод, который работает с любыми типами студентов.


```python
class Student:
    """Класс для обычного студента.
    
    Атрибуты:
        name (str): Имя студента.
        _age (int): Возраст студента.        
    """

    def __init__(self, name:str, age:int):
        """Конструктор класса Student.

        Args:
            name (str): Имя студента.
            age (int): Возраст студента.
        """
        self.name = name
        self._age = age

    def get_info(self):
        """Возвращает информацию о студенте.

        Returns:
            str: Строка формата "Студент {name}, возраст {age}"
        """
        return f'Студент {self.name}, возраст {self._age}'
    
    def __str__(self):
        """Красивый вывод объекта в виде строки.

        См. get_info()
        """
        return self.get_info()
    
    def __repr__(self):
        """Строковое представление объекта.

        Returns:
            str: Строка формата "Student(name={self.name}, age={self._age})"
        """
        return f'Student(name={self.name}, age={self._age})'


class GraduateStudent(Student):
    """Класс для аспиранта, наследуюется от Student.
    
    Атрибуты:
        name (str): Имя студента.
        _age (int): Возраст студента.  
        _research_topic (str): Тема исследования.
        __publications (int): Количество публикаций.    
    """

    def __init__(self, name:str, age:int, research_topic:str):
        """Конструктор класса GraduateStudent.

        Args:
            name (str): Имя студента.
            age (int): Возраст студента.
            research_topic (str): Тема исследования.
        """
        super().__init__(name, age)
        self._research_topic = research_topic
        self.__publications = 0
    
    @property
    def publications(self):
        """Возвращает количество публикаций.

        Returns:
            int: Количество публикаций.
        """
        return self.__publications
    
    def add_publication(self):
        """Увеличивает количество публикаций на 1."""
        self.__publications += 1
    
    def get_info(self):
        """Возвращает информацию о студенте.

        Returns:
            str: Строка формата "Студент {name}, возраст {age}, тема: {research_topic}"
        """        
        return f'Студент {self.name}, возраст {self._age}, тема: {self._research_topic}'

    def __repr__(self):
        """Строковое представление объекта.

        Returns:
            str: Строка формата "GraduateStudent(name={self.name}, age={self._age}, research_topic={self._research_topic})"
        """
        return f'GraduateStudent(name={self.name}, age={self._age}, research_topic={self._research_topic})'


class StudentManager:
    """Класс для менеджера для работы с коллекцией студентов."""

    def __init__(self):
        """Конструктор класса StudentManager. Создаёт пустой список студентов."""
        self._students_list = []
    
    def add_student(self, student:Student):
        """Добавляет студента в коллекцию.

        Args:
            student (Student): Объект студента или его наследника.
        """
        self._students_list.append(student)

    def print_all_info(self):
        """Выводит информацию о всех студентах в коллекции."""
        print('\nИнформация о всех студентах:')
        for i, student in enumerate(self._students_list, start=1):
            print(f'{i}: {student}')
        print()
    
    def __len__(self):
        """Возвращает количество студентов.

        Returns:
            int: Количество студентов в коллекции.
        """
        return len(self._students_list)
    
    def __getitem__(self, i):
        """Возвращает студента из коллекции по индексу.

        Args:
            i (int): Индекс.

        Returns:
            Student: Студент.
        """
        return self._students_list[i]
    
    def __repr__(self):
        """Строковое представление объекта.

        Returns:
            str: Cтрока вида StudentManager([...]).
        """
        return f'StudentManager({self._students_list})'
```

```python
# Образец использования:
manager = StudentManager()

student1 = Student('Анна', 20)
student2 = GraduateStudent('Петр', 25, 'AI')

manager.add_student(student1)
manager.add_student(student2)

print('Всего студентов:', len(manager))  # Должно вывести: 2
print('Студент 1:', manager[0])    # Должен вывести первого студента

# Полиморфный метод, который выводит информацию о всех студентах
manager.print_all_info()

print('Строковое представление менеджера:', manager)
```
